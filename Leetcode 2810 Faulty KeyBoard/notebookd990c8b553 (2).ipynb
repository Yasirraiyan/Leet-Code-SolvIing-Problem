```python


import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Part 1: Reverse Visualization
# -----------------------------

# Example string
original = "hello"
reversed_steps = []

# Build reversed string step-by-step
temp = ""
for i in range(len(original) - 1, -1, -1):
    temp += original[i]
    reversed_steps.append(temp)

# Plot reverse visualization
plt.figure(figsize=(10, 6))
for i, step in enumerate(reversed_steps):
    plt.text(i, 0.5, f'Step {i+1}: "{step}"', fontsize=12, color='purple')

plt.title('Step-by-Step Reverse Visualization of "hello"', fontsize=14, fontweight='bold')
plt.axis('off')
plt.subplots_adjust(left=0.05, right=0.95)  # fix margins
plt.show()  # Show the plot
plt.savefig("reverse_visualization.jpeg", format='jpeg')
plt.close()

# -----------------------------------
# Part 2: Time & Space Complexity Plot
# -----------------------------------

# Input size from 1 to 100
input_sizes = np.arange(1, 101)
time_complexity = input_sizes           # O(n)
space_complexity = input_sizes          # O(n)

# Plot time and space complexity
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, time_complexity, label='Time Complexity O(n)', color='green', linewidth=2)
plt.plot(input_sizes, space_complexity, label='Space Complexity O(n)', color='blue', linestyle='--', linewidth=2)

plt.title('Time & Space Complexity of reverse() Function', fontsize=14, fontweight='bold')
plt.xlabel('Input Size (n)', fontsize=12)
plt.ylabel('Operations / Memory', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()  # Show the plot
plt.savefig("reverse_complexity_graph.jpeg", format='jpeg')
plt.close()
```
